Data Preprocessing ----> Data cleaning

```python
df['order_date'] = pd.to_datetime(df['order_date'], format = 'mixed')
df['ship_date'] = pd.to_datetime(df['ship_date'], format = 'mixed')
```

```python
df.drop(columns = ['customer_country'], inplace = True)
```

```python
def shipping_type(days_for_shipment_scheduled):

    if days_for_shipment_scheduled == 1:
        return "First Class"
    elif days_for_shipment_scheduled == 2:
        return "Second Class"
    elif days_for_shipment_scheduled == 4:
        return "Standard Class"
    elif days_for_shipment_scheduled == 0:
        return "Same Day"
    else:
        return ""

df['shipping_type'] = df['days_for_shipment_scheduled'].apply(shipping_type)
```
